```python
df['datetime'] = pd.to_datetime(df['Date']+" "+df['Time'])
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['Month_name']=df['datetime'].dt.month_name()
df['Year_Month']=df['year'].apply(str)+' '+df['Month_name'].apply(str)
df['day'] = df['datetime'].dt.day
df['weekday'] = df['datetime'].dt.weekday
df['hour'] = df['datetime'].dt.hour
df = df.drop(['Date'], axis=1)

df.head(5)
```

```python
df.groupby('month')['Item'].count().plot(kind='bar')
```

```python
df['monthly']=pd.to_datetime(df['datetime']).dt.to_period('M')
monthlyTransactionForItem=df[['monthly','Transaction','Item']].groupby(['monthly','Item'],as_index=False).count().sort_values(by='monthly')
monthlyTransactionForItem.set_index('monthly',inplace=True)

monthlyTransactionForItem.head(35)
```

```python
coffeesalesmonthly=monthlyTransactionForItem[monthlyTransactionForItem['Item']=='Coffee'].plot(kind='bar')
coffeesalesmonthly.set_xlabel("Coffee Sales Monthly")

plt.ylabel('Transaction', fontsize=16)
plt.xlabel('Month', fontsize=16)
plt.title('Monthly Coffee Sales', fontsize=16);
```

```python
df.groupby('hour')['Item'].count().plot(kind='bar')
```

```python
df['hourly']=pd.to_datetime(df['datetime']).dt.to_period('H')
hourlyTransactionForItem=df[['hourly','Transaction','Item']].groupby(['hourly','Item'],as_index=False).count().sort_values(by='hourly')
hourlyTransactionForItem.set_index('hourly',inplace=True)
```

```python
coffeesaleshourly=hourlyTransactionForItem[hourlyTransactionForItem['Item']=='Coffee'].plot(kind='bar',figsize=(20,20))
coffeesaleshourly.set_xlabel("Coffee Sales hourly")

plt.ylabel('Transaction', fontsize=16)
plt.xlabel('hourly', fontsize=16)
plt.title('hourly Coffee Sales', fontsize=16);
```
